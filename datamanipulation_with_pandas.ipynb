Data Manipulation and Data analysis are the key task in any data science or Data Analytics Project. Pandas provide a wide range of functions for Data manipulation and Analysis, making it easier to clean, transform and extractinsights from data.

```python
#Handling Missing values
df.isnull()#True means there is a missing value, but it's hard to read all
```

```python
df.isnull().sum()
```

```python
df_filled= df.fillna(0) #I twil fill the null vales with 0
```

```python
#filling missing values by the mean of the coulmn
df['Series_title_5_fill']= df['Series_title_5'].fillna(df['Series_title_5'].mean())
df
```

```python
#Renaming columns
df= df.rename(columns={'Series_reference': 'Reference'})
df.head()
```

```python
df.dtypes
```

```python
#Changing Data types

df['Magnitude_new']=df['Magnitude'].fillna(df['Magnitude'].mean()).astype(float)
```

```python
df.head()
```

```python
#Data Aggregating and grouping
grouped_mean= df.groupby('Group')['Data_value'].mean()
print(grouped_mean)
```
