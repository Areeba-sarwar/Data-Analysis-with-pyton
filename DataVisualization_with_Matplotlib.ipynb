# Matplotlib:

Matplotlib is a powerful plotting library for python that enables the creation of static, animated and interactive visualizations. It is widely used for Data Visualization in data sceince and Data Analytics

```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,3,4,5]
y= [22,34,46,56,60]

#We want to draw a line graph
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('This is my line plot')
plt.show()
```

```python
x= [1,2,3,4,5]
y= [22,34,46,56,60]

#Create a customized Line plot
plt.plot(x,y, color= 'red', linestyle= '-.', marker= 'o' , linewidth= 3, markersize= 9)
plt.grid(True)
```

```python
#Mutiple plots
##Sample data

x= [1,2,3,4,5]
y1= [1,4,6,8,10]
y2=[30,35,40,45,50]

plt.Figure(figsize= (5,9))
plt.subplot(2,2,1)
plt.plot(x,y1,color='Green')
plt.title('Plot 1')

plt.subplot(2,2,2)
plt.plot(x,y2,color='Green')
plt.title('Plot 2')


plt.subplot(2,2,3)
plt.plot(x,y2,color='Green')
plt.title('Plot 3')


plt.subplot(2,2,4)
plt.plot(x,y2,color='Green')
plt.title('Plot 4')

```

```python
#Creating Bar plot
categories= ['A','B','C','D','E']
Values= [4,2,5,8,3]

plt.bar(categories,Values,color='green')
```

## Histogram:

Histograms are used to represent the distribution of a dataset. They divide the data into bins and counts the data points in each bin

```python
## Sample Data

data= [1,2,2,2,3,4,5,6,2,3,4,4,2,3,4,4,2]

plt.hist(data, bins=5, color='orange',edgecolor= 'black')
```

```python
##Creating a scatter plot
x= [1,2,3,4,5]
y=[2,3,4,5,6]

plt.scatter(x,y, color= 'red', marker= 'x')
```

```python
##Pie Chart
labels= ['A', 'B', 'C', 'D', 'E']
sizes= [30,20,40,50,35]
colors= ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode= (0.1,0,0.2,0,0) ##move out the first and third label with given distance

#Create a pie chaart
plt.pie(sizes,explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True)
```
